```python
# Remove Duplicates from a List
lst = [1, 2, 2, 3, 4, 4, 5]
unique_lst = list(set(lst))
print(unique_lst)
```

```python
# Remove Duplicates from a List
lst = [1, 2, 2, 3, 4, 4, 5]

unique_lst = []
for x in lst:
    if x not in unique_lst:
        unique_lst.append(x)

print(unique_lst)
```

```python
# Find Duplicates
lst = [1, 2, 3, 2, 4, 5, 1]

duplicates = []
seen = set()

for x in lst:
    if x in seen:
        duplicates.append(x)
    else:
        seen.add(x)

print(duplicates)
```

**2- Sets**

```python
"""
UnOrderd [not indexed] >> Handled by membership (in, not in)
sets contain only immutable data type (int, str, tuple)
items are unique
"""
```

```python
a = {1, 2, 3, 4}  # set
b = set([2, 3, 5])

3 in a

'''
a.add(5)
a.remove(2)     # يشيب العنصر ويطلع ايرور لو مش موجود
a.discard(2)    # يشيل العنصر ولو مش هيطلع ايرور
a.pop()         # يشيل عنصر عشوائي
a.clear()       # يفضي set
'''
```

```python
lst = [1, 3, 4, 3]
if len(lst) != len(set(lst)):
    print("Has duplicates")
```

```python
a = {1, 2, 3}
b = {2, 3, 4}

a.union(b)       # {1,2,3,4}  → a | b
a.intersection(b) # {2,3}    → a & b
a.difference(b)   # {1}      → a - b
a.symmetric_difference(b) # {1,4}
```

```python
a = [1,2,3]
b = [2,3,4]

common = set(a) & set(b)
only_a = set(a) - set(b)
print(common, only_a)
```

**3- Dictionaries**

```python
"""
ordered in recent Python versions
keys has to be only immutable data type (int, str, tuple)
"""
```

```python
d = {'a': 1, 'b': 2, 'c': 3}

# طريقة تانية
d2 = dict(a=1, b=2, c=3)

# print(d['x']) → KeyError
print(d.get('x'))        # None
d.get('x', 0)     # 0 → default value
```

```python
d.keys()       # dict_keys(['a','b','c'])
d.values()     # dict_values([1,2,3])
d.items()      # dict_items([('a',1),('b',2),('c',3)])

d.pop('b')     # يحذف ويرجع القيمة
d.popitem()    # يحذف آخر عنصر
d.clear()      # يفضي dict
d.update({'d':4, 'e':5}) # يضيف / يحدث
```

```python
s = "banana"
freq = {}
for c in s:
    freq[c] = freq.get(c, 0) + 1
# {'b':1, 'a':3, 'n':2}
```

```python
nums = [1, 2, 3, 2, 4, 2, 5, 3]
freq = {}

for num in nums:
    freq[num] = freq.get(num, 0) + 1

most_common_element = max(freq, key=freq.get)
frequency = freq[most_common_element]

print("Most Frequent Element:", most_common_element)
print("Frequency:", frequency)
```

```python
d = {'a': 1, 'b': 2, 'c': 3}
for key, val in d.items():
    print(key, val)
print()
for key in d:
    print(key, d[key])
```

```python
nums = [5, 2, 9, 1]

for i in range(len(nums)):
    for j in range(i + 1, len(nums)):
        if nums[i] > nums[j]:
            nums[i], nums[j] = nums[j], nums[i]

print(nums)
```

```python
# Recovering a Small String
t = int(input())

for _ in range(t):
    n = int(input())
    for a in range(1, 27):
        for b in range(1, 27):
            c = n - a - b
            if 1 <= c <= 26:
                print(
                    chr(ord('a') + a - 1) +
                    chr(ord('a') + b - 1) +
                    chr(ord('a') + c - 1)
                )
                break
        else:
            continue
        break
```

```python
# Count Vowels
s = "hello world"
count = 0

for ch in s:
    if ch.lower() in "aeiou":
        count += 1

print(count)
```

```python
# Count Vowels
from collections import Counter

s = "hello world"
c = Counter(s.lower())

vowels = "aeiou"
count = sum(c[v] for v in vowels)

print(count)
```
