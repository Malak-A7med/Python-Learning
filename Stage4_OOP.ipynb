Object-Oriented Programming (OOP) is a programming approach where a program is divided into objects.
Each object represents a real-world entity and contains:

Data → Attributes (or Properties)

Behavior → Methods (or Functions)


=> In simple words:
Object = Data + Methods

Example:
  Student

Object:

An Object is an instance of a class.
It is the real thing created from the class.
Example:
  student1

Class → Blueprint

**Python doesn’t have real private variables, only name mangling**

```python
# Encapsulation
class person:
  def __init__(self, name, id):
    self.person_name = name
    self.__id = id
    print(f"Welcome {self.person_name}")

p1 = person("Malak", 3)
print(p1._person__id)
```

---

**Abstraction:**

make structure
and let the implementation to the child

```python
# What to do, not how to do

from abc import ABC, abstractmethod

class Model(ABC):

    @abstractmethod
    def train(self):
        pass
"""
You cannot create an object from Model
Any subclass must implement train()
"""

class LinearRegression(Model):
    def train(self):
        print("Training Linear Regression")

class NeuralNetwork(Model):
    def train(self):
        print("Training Neural Network")

models = [LinearRegression(), NeuralNetwork()]
for model in models:
    model.train()
```

| Abstraction          | Encapsulation         |
| -------------------- | --------------------- |
| What an object does  | How data is protected |
| Interface / Contract | Data hiding           |
| `abstractmethod`     | private variables     |

**=> Polymorphism**

**=> One interface, many behaviors**

```python
# with Inheritance
class Shape:
    def area(self):
        pass

class Rectangle(Shape):
    def area(self):
        return 10 * 5

class Circle(Shape):
    def area(self):
        return 3.14 * 7 ** 2

shapes = [Rectangle(), Circle()]

for shape in shapes:
    print(shape.area())
```

```python
class LinearRegression:
    def train(self):
        print("Training Linear Regression")

class NeuralNetwork:
    def train(self):
        print("Training Neural Network")

models = [LinearRegression(), NeuralNetwork()]
for model in models:
    model.train()
```
