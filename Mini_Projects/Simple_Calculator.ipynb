```python
def calculator():
    print("Welcome to your manual calculator! ")
    print("Type 'exit' to quit.")

    while True:
        expression = input("\nEnter The Equation: ")

        if expression.lower() == "exit":
            print("Goodbye!")
            break

        operator = ""
        for op in "+-*/":
            if op in expression:
                operator = op
                break

        if operator == "":
            print("No operator found! Use +, -, *, /")
            continue

        num1_str, num2_str = expression.split(operator)

        try:
            num1 = float(num1_str)
            num2 = float(num2_str)

            if operator == "+":
                result = num1 + num2
            elif operator == "-":
                result = num1 - num2
            elif operator == "*":
                result = num1 * num2
            elif operator == "/":
                if num2 != 0:
                    result = num1 / num2
                else:
                    result = "Error: Division by zero!"

            print("Result:", result)

        except ValueError:
            print("Invalid numbers! Try again.")

calculator()
```
