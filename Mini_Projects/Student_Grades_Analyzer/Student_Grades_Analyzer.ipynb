```python
"""
Goal:
You have a file with student names and their scores.
Your app will read this file, calculate averages,
determine grades, and save a summary to another file.
"""
```

```python
# Step 1: Read data from a file
def read_data(filename):
    students = []
    with open(filename, 'r') as file:
        lines = file.readlines()
        headers = lines[0].strip().split(",")  # ['Name', 'Math', 'English', 'Science']
        for line in lines[1:]:
            parts = line.strip().split(",")
            student = {
                "Name": parts[0],
                "Math": int(parts[1]),
                "English": int(parts[2]),
                "Science": int(parts[3])
            }
            students.append(student)
    return students

# Step 2: Process / Analyze data
def calculate_grades(students):
    for student in students:
        avg = (student['Math'] + student['English'] + student['Science']) / 3
        student['Average'] = round(avg, 2)
        if avg >= 90:
            student['Grade'] = 'A'
        elif avg >= 80:
            student['Grade'] = 'B'
        elif avg >= 70:
            student['Grade'] = 'C'
        elif avg >= 60:
            student['Grade'] = 'D'
        else:
            student['Grade'] = 'F'
    return students

# Step 3: Write output to a file
def write_summary(students, output_file):
    with open(output_file, 'w') as file:
        file.write("Name,Math,English,Science,Average,Grade\n")
        for s in students:
            line = f"{s['Name']},{s['Math']},{s['English']},{s['Science']},{s['Average']},{s['Grade']}\n"
            file.write(line)

# Main function
def main():
    students = read_data('students.txt')
    students = calculate_grades(students)
    write_summary(students, 'summary.txt')

    from google.colab import files
    files.download("summary.txt")

    print("Summary saved to summary.txt ✅")

if __name__ == "__main__":
    main()
```
