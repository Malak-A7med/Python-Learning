```python
class Task:
  def __init__(self, title, desc = ""):
      self.title = title
      self.description = desc
      self.status = "Pending" #default

  def mark_done(self):
      self.status = "Done"

  def update_title(self, new_title):
      self.title = new_title

  def update_description(self, new_desc):
      self.description = new_desc

  def __str__(self):
      return f"{self.title} - {self.description} [{self.status}]"
```

```python
# Add, Remove, Update, Show
class Task_manager:
  def __init__(self):
    self.tasks = []
    # the list here has objects of The Class Task not strings, for example
    """
    Task("Study", "Python OOP")
    Task("Workout", "Leg day")
    """

  def add_task(self, task):
    self.tasks.append(task) # task here an object from Task class
    print(f"Task {task.title} added! ")

  def remove_task(self, title):
    for task in self.tasks:
        if task.title == title:
            self.tasks.remove(task)
            print(f"Task '{title}' removed!")
            return
    print(f"Task '{title}' not found!")

  def update_task(self, title, new_title=None, new_description=None, mark_done=False):
    for task in self.tasks:
        if task.title == title:
            if new_title:
                task.update_title(new_title)
            if new_description:
                task.update_description(new_description)
            if mark_done:
                task.mark_done()
            print(f"Task '{title}' updated!")
            return
    print(f"Task '{title}' not found!")

  def show_tasks(self, status = None):
    for t in self.tasks:
      if status is None or t.status == status:
        print(t)
        # it;s like saying => print(object)
        # it has been implemented cus __str__

```

```python
def main():
  manager = Task_manager()

  while True:
    print("\n=== To-Do List Menu ===")
    print("1. Add Task")
    print("2. Remove Task")
    print("3. Update Task")
    print("4. Show All Tasks")
    print("5. Show Pending Tasks")
    print("6. Show Done Tasks")
    print("7. Exit")

    choice = input("Enter your choice: ")

    if choice == "1":
      title = input("Enter task title: ")
      desc = input("Enter description (optional): ")
      manager.add_task(Task(title, desc)) # Task(title, desc) => this make "Task object from Task class"

    elif choice == "2":
      title = input("Enter the task title to remove: ")
      manager.remove_task(title)

    elif choice == "3":
      title = input("Enter task title to update: ")
      new_title = input("Enter new title (press enter to skip): ")
      new_desc = input("Enter new description (press enter to skip): ")
      done = input("Mark as done? (y/n): ").lower() == 'y' # if y ==> (done = True)
      manager.update_task(title, new_title or None, new_desc or None, done)

    elif choice == "4":
      manager.show_tasks()

    elif choice == "5":
      manager.show_tasks(status = "Pending")

    elif choice == "6":
      manager.show_tasks(status = "Done")

    elif choice == "7":
      print("Good Bye!")
      break

    else:
      print("Invalid choice, try again.")

if __name__ == "__main__":
    main()
```
