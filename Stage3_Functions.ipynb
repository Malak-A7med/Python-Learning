==> Using *args and **kwargs in Python

```python
# --------------------------
# *args example
# --------------------------
def print_args(*args):
    # *args collects all extra positional arguments into a tuple
    # This allows the function to accept any number of arguments
    # You can loop over them, index them, or do any tuple operations
    for a in args:
        print(a)

print_args(1, 2, 3)
print()
print_args(1, 2)
```

```python
# --------------------------
# **kwargs example
# --------------------------
def print_kwargs(**kwargs):
    # **kwargs collects all extra keyword arguments(kwargs) into a dictionary
    # This allows the function to accept named arguments that are not fixed
    # You can loop over them using items() to get key-value pairs
    for key, value in kwargs.items():
        print(key, value)

print_kwargs(name="Malak", age=20)
print_kwargs(city="Cairo", country="Egypt", hobby="AI")
```

```python
# --------------------------
# Combining *args and **kwargs
# --------------------------
def combined_example(*args, **kwargs):
    # Here, *args collects positional arguments
    # and **kwargs collects keyword arguments
    print("Positional arguments:", args)
    print("Keyword arguments:", kwargs)

combined_example(1, 2, 3, name="Malak", age=20)
```

```python
def fibonacci(n):
    if n <= 1:
        return n
    return fibonacci(n-1) + fibonacci(n-2)
print(fibonacci(6))
```

```python
# Binary Search
def binary_search(arr, target):
  l, r = 0, len(arr) - 1
  while l <= r:
    mid = (l + r) // 2
    if arr[mid] == target:
      return mid + 1  # return the index of the target
    elif arr[mid] < target:
      l = mid + 1
    else:
      r = mid - 1
  return -1

arr = list(map(int, input().split()))
target = int(input())
print(f"The index of the target is: {binary_search(arr, target)}")

```

```python
# Recursive Factorial
def fact(n):
  if n in (0, 1):
    return 1
  else:
    return n * fact(n - 1)

num = int(input("Enter a number: "))
print(f"{num}! = {fact(num)}")
```

```python
# max_subarray_sum
nums = [-2,1,-3,4,-1,2,1,-5,4]

cur_sm, mx_sm = nums[0], nums[0]
for i in range(1, len(nums)):
  cur_sm = max(nums[i], cur_sm + nums[i])
  mx_sm = max(cur_sm, mx_sm)

print(mx_sm)  # Output: 6
```

```python
# Word Count
s = input().lower().split()
cnt = {}
for word in s:
  cnt[word] = cnt.get(word, 0) + 1

for i, j in cnt.items():
  print(f"{i} : {j}")
```

```python
# Most Frequent Word
s = input().lower().split()
cnt = {}
for word in s:
  cnt[word] = cnt.get(word, 0) + 1

most_word = max(cnt, key = cnt.get)
print(most_word, cnt[most_word])

smallest_word = min(cnt, key = cnt.get)
print(smallest_word, cnt[smallest_word])
```
