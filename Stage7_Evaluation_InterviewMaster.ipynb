```python
import math
n = int(input())
x = 0
if n <= 1:
  print("False")
elif n > 1:
  for i in range(2, int(math.sqrt(n) + 1)):
    if n % i == 0:
      x = 1
      break
  print("True" if x == 0 else "False")
```

```python
# Euclidean Algorithm
def gcd(a, b):
    while b != 0:
        a, b = b, a % b
    return a

a = int(input("Enter first number: "))
b = int(input("Enter second number: "))

print("GCD:", gcd(a, b))
```

```python
def gcd(a, b):
    if b == 0:
        return a
    else:
        return gcd(b, a % b)

a = int(input("Enter first number: "))
b = int(input("Enter second number: "))

print("GCD:", gcd(a, b))
```

```python
def min_swaps(s, ch):
    pos = [i for i, c in enumerate(s) if c == ch]
    if not pos:
        return 0
    m_i = len(pos) // 2
    median = pos[m_i]
    t = 0
    for i, p in enumerate(pos):
        t += abs(p - (median - m_i + i))
    return t

t = int(input())
for _ in range(t):
    input()
    s = input().strip()
    print(min(min_swaps(s, 'a'), min_swaps(s, 'b')))
```
