```python
# Simple Calculator Program
equation = input()
num1, num2, op = '', '', ''
for c in equation:
  if c in '*/+-':
    op = c
  elif op == '':
    num1 += c
  else:
    num2 += c

a, b = map(int, [num1, num2])
if op == '+':
  print(a+b)
elif op == '-':
    print(a - b)
elif op == '*':
    print(a * b)
elif op == '/':
    if b != 0:
        print(a / b)
    else:
        print("Error: Division by zero")
```

```python
#	Bubble sort
n = int(input())
arr = list(map(int, input().split()))

for i in range(n):
    for j in range(0, n - i - 1):
        if arr[j] > arr[j + 1]:
            arr[j], arr[j + 1] = arr[j + 1], arr[j]

print(*arr)
```

```python
# Maximum number
n = int(input())
arr = list(map(int, input().split()))

mx = arr[0]
for x in arr:
    if x > mx:
        mx = x

print(mx)
```

```python
# Prime Check
n = int(input())
if n < 2:
    print("Not Prime")
else:
  flag = 1
  for i in range(2, int(n**0.5) + 1):
    if n % i == 0:
      flag = 0
      break
if flag:
  print("Prime")
else:
  print("Not Prime")
```

```python
# Max Subsequence
n = int(input())
s = input()
last = ''
ans = ''
count = 0

for c in s:
    if c != last:
        count += 1
        ans += c
        last = c

print(count, ans)
```

```python
# Max Substring
n = int(input())
s = input()

max_len = 0
cur_len = 0
last = ''
cur_sub = ''
max_sub = ''

for c in s:
    if c != last:
        cur_len += 1
        cur_sub += c
    else:  # start new substring
        cur_len = 1
        cur_sub = c
    last = c

    if cur_len > max_len:
        max_len = cur_len
        max_sub = cur_sub

print(max_len, max_sub)
```
